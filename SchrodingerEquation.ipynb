```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define constants
L = 10.0  # Length of the box
N = 1000  # Number of points
x = np.linspace(0, L, N)  # x values
dx = x[1] - x[0]  # Difference between successive x values
```

```python
# Define the Laplacian operator as a matrix
Laplacian = (-2.0*np.diag(np.ones(N)) + np.diag(np.ones(N-1),1) + np.diag(np.ones(N-1),-1)) / (dx**2)
```

```python
# Define the Hamiltonian operator
Hamiltonian = -(1/2) * Laplacian
```

```python
# Compute the eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(Hamiltonian)
```

```python
# Sort the eigenvalues and eigenvectors
sorted_indices = np.argsort(eigenvalues)
eigenvalues = eigenvalues[sorted_indices]
eigenvectors = eigenvectors[:, sorted_indices]
```

```python
# Plot the lowest few energy levels
for i in range(4):
    plt.plot(x, eigenvectors[:, i] + eigenvalues[i])
```

```python
plt.title("Particle in a Box")
plt.xlabel("Position")
plt.ylabel("Energy")
plt.grid(True)
plt.show()
```
